```python
def neighbourhood(df1,neigh):
    new_df = df1[df1['host_neighbourhood']=='Le Plateau']
    new1 = new_df.groupby('room_type').sum('price')
    new1.reset_index(inplace=True)
    fig_1 =px.bar(new1,x = 'room_type',y = 'price',title = "Room type and Price",
                color_discrete_sequence=px.colors.sequential.Rainbow_r,height = 650, width = 600)
    fig_1.show()

    new2 = new_df.groupby('property_type').sum('accommodates')
    new2.reset_index(inplace=True)

    fig_2 = px.bar(new2,x = 'property_type',y = 'accommodates',title = "Property type and accommodates",
                        color_discrete_sequence=px.colors.sequential.Cividis_r,height = 650, width = 600 )
    fig_2.show()
```

```python
def availability(df,val):
    if val == 30:
        figure_1 = px.pie(df,values = 'availability_30',names= 'country',title = "COUNTRY AND AVAILABILITY FOR 30 DAYS")
        figure_1.show()
    elif val==60:
        figure_2 = px.pie(df,values='availability_60', names = 'country',title = "COUNTRY AND AVAILABILITY FOR 60 DAYS")
        figure_2.show()
    elif val == 90:
        figure_3 = px.pie(df,values='availability_90', names = 'country',title = "COUNTRY AND AVAILABILITY FOR 90 DAYS")
        figure_3.show()
    elif val == 365:
        figure_4 = px.pie(df,values='availability_365', names = 'country',title = "COUNTRY AND AVAILABILITY FOR 365 DAYS")
        figure_4.show()
```
